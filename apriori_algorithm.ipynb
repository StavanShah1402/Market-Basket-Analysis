```python
df=data.drop(['Unnamed: 0'],axis=1)
df.head()
```

```python
df[df['item_name']=="Focaccia, Spicy Meatballs"]
```

```python
df=df.groupby(['order_id','item_name'])['quantity'].sum().unstack().reset_index().fillna(0).set_index('order_id')
df.head()
```

```python
def encode_quantity(x):
    if x<=0:
        return 0
    if x>0:
        return 1
df=df.applymap(encode_quantity)
df.head()
```

```python
df.describe()
```

#Filtering dataset with items greater than 1

```python
df_filter=df[(df>0).sum(axis=1)>=2]
df_filter
```

#Applying the apriori algorithm

```python
from mlxtend.frequent_patterns import apriori

frequent_items=apriori(df_filter,min_support=0.03,use_colnames=True).sort_values('support',ascending=False).reset_index(drop=True)
frequent_items['length']=frequent_items['itemsets'].apply(lambda x:len(x))
frequent_items
```

```python
from mlxtend.frequent_patterns import association_rules

df=association_rules(frequent_items,metric='lift',
                  min_threshold=1).sort_values('lift',ascending=False).reset_index(drop=True)
```
